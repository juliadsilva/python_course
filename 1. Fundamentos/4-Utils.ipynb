#### Validação
A instrução `assert` é uma ferramenta de depuração (debug) que ajuda a encontrar erros no código durante a fase de desenvolvimento. Ela serve para testar se uma condição é verdadeira; se não for, o Python interrompe a execução do programa e levanta um erro.

1. (Capitalize - HackerRank) Você deve garantir que o primeiro e o último nome das pessoas comecem com uma letra maiúscula em seus passaportes. Por exemplo, alison heck deve ser escrito corretamente como Alison Heck.

Considere a função built-in `capitalize()` e faça uma validação de input para validar se este é uma string.

```python
def capitalize_name(s):
    assert isinstance(s, str) , "input is not a string"
    new_s = [(lambda x : x.capitalize()) (x) for x in s.split()]
    return " ".join(new_s)
```

```python
new_name = capitalize_name("carolina kamada")
print(new_name)
```

```python
capitalize_name(23)
```

#### Regex
Atuam como uma busca e substituição. Elas permitem que você encontre, valide ou extraia padrões complexos em textos — como e-mails, números de telefone, CPFs ou datas — usando uma linguagem de símbolos específica.
<br> 

Para criar um padrão, você combina caracteres normais com símbolos especiais:
|Padrão|Significado|
|-------|----------|
|^|Nega a expressão OU marca o início de uma string|
|$|Marca o final de uma string|
|+|Prolonga o caractere anterior|
|.|Substitui qualquer caractere (apenas um)|
|[A-Z]|Procura qualquer caractere maiúsculo de A a Z|
|[a-z]|Procura qualquer caractere minúsculo de a a z|
|[0-9]|Procura qualquer dígito de 0 a 9|
|[125]|Procura os dígitos 1, 2 ou 5|
|[^0-9]|O ^ nega a expressão a seguir, logo, procura tudo que não for número|
|[abc]|"Procura os caracteres a, b ou c"|
|[^A-Z]|Procura tudo que não for letra maiúscula|
|[a-zA-Z]|Procura tudo que for letra|
|\w|"Procura caracteres, exceto os especiais"|
|\d|Também procura qualquer dígito de 0 a 9|

2. Defina uma função para validar se um endereço de e-mail é válido. Considere:
- Uso da biblioteca `re`
- O seguinte padrão: <>@<>.com

```python
import re
```

```python
def valida_email(s):
    if re.match("^[a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+.com$", s):
        print("email válido")
    else:
        print("email não válido")
```

```python
valida_email("caroll.zambelli@gmail.com")
```

```python
valida_email("caroll.zambelligmail.com")
```

#### Log e Gerenciamento de Exceção

Diferente do comando `print()`, que apenas joga dados no console, o logging permite direcionar mensagens para arquivos, definir níveis de severidade e filtrar o que é realmente importante.

3. Dado uma string, fazer uma função que conta as letras de cada palavra, utilizando a função `lambda` com `map` . O formato de saída da função deve ser um dicionário.

https://medium.com/@habbema/logging-em-python-518b162de84d

```python
import logging
```

```python
def word_counter(s):
    try:
        results = map(lambda w: [w, len(w)] , s.split())
        print(dict(results))
    except Exception as expection_error:
        #print(expection_error)
        logging.error(expection_error)
        
```

```python
word_counter('Impacta offers the best data courses in Brazil')
```

```python
logging.basicConfig(filename="app.log",
                    format='%(asctime)s - %(levelname)s - %(message)s')
```

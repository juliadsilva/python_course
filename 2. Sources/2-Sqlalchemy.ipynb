### SQLAlchemy

Diferente do SQLite, que é um banco de dados real, o SQLAlchemy é um ORM (Object Relational Mapper) uma ferramenta de software que atua como uma "ponte" entre o seu código e o banco de dados.

|Aspecto | SQL Puro (Drivers) | SQLAlchemy|
----------|------------------|-------------
|Curva de Aprendizado | Baixa (se já souber SQL) | Média/Alta|
|Velocidade de Execução | Máxima | Boa (com leve overhead)|
|Portabilidade | Difícil (SQL varia entre bancos) | Excelente (troca o banco via config)|
|Segurança | Manual (risco de SQL Injection) | Automática (muito seguro)|

```python
from sqlalchemy import create_engine, MetaData, Table, select, text
```

```python
engine = create_engine("sqlite:///aula02.db")
```

```python
sql = text("SELECT * from books")
```

```python
with engine.connect() as conn:
    result = conn.execute(sql)
    print(result.fetchall())
```

```python
metadata = MetaData()
```

```python
t = Table("books", metadata, autoload_with=engine)
```

```python
t.columns
```

```python
[col.key for col in t.columns]
```

```python
sql2 = select(t.columns.title, t.c.author_id).where(t.c.author_id == '2')
print(sql2)
```

```python
with engine.connect() as conn:
    result = conn.execute(sql2)
    print(result.fetchall())
```

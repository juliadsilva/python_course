
## Requests

A biblioteca requests é um pacote Python que simplifica o trabalho com requisições HTTP.

```python
import pandas as pd
import requests
```

### Explorando o retorno do método get

Métodos básicos de requisição HTTP:
- GET: Solicita a representação de um recurso.
- POST: Envio de dados para serem processados por um recurso.
- PUT: Atualização de um recurso existente.
- DELETE: Exclusão de um recurso.

Status Codes
- 2xx para sucesso
- 3xx para redirecionamentos
- 4xx para erros do cliente
- 5xx para erros do servidor

```python
api_url = "https://randomuser.me/api/?results=5"
```

```python
reponse_api = requests.get(api_url, timeout=10)
reponse_api
```

```python
print(reponse_api.cookies)
print(reponse_api.connection)
print(reponse_api.content)
```

```python
response = reponse_api.json()
response
```

```python
data = response['results']
df = pd.json_normalize(data)
df.head()
```

```python
cols = [s.replace('.', '_') for s in df.columns]
df.columns = cols
df.columns
```
